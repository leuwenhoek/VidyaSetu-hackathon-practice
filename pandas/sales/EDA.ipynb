## Feature Extraction 

### Duration of

```python
from datetime import datetime
def duration(d1,d2):

    date1 = datetime.strptime(d1, "%m/%d/%Y")
    date2 = datetime.strptime(d2, "%m/%d/%Y")

    gap = date2 - date1
    return gap.days

df["Duration"] = df.apply(lambda x: duration(x['Order Date'],x['Ship Date']),axis=1)
```

### Order Hash

```python
df['Order Hash'] = hash(lambda x: x['Order ID'])
df
```
